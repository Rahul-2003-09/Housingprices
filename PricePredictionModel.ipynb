```python
def heatMap(df, mirror):

   # Create Correlation df
   corr = df.corr()
   # Plot figsize
   fig, ax = plt.subplots(figsize=(15, 6))
   # Generate Color Map
   colormap = sns.diverging_palette(220, 10, as_cmap=True)
   
   if mirror == True:
      #Generate Heat Map, allow annotations and place floats in map
      sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f")
      #Apply xticks
      plt.xticks(range(len(corr.columns)), corr.columns);
      #Apply yticks
      plt.yticks(range(len(corr.columns)), corr.columns)
      #show plot

   else:
      # Drop self-correlations
      dropSelf = np.zeros_like(corr)
      dropSelf[np.triu_indices_from(dropSelf)] = True
      # Generate Color Map
      colormap = sns.diverging_palette(220, 10, as_cmap=True)
      # Generate Heat Map, allow annotations and place floats in map
      sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f", mask=dropSelf)
      # Apply xticks
      plt.xticks(range(len(corr.columns)), corr.columns);
      # Apply yticks
      plt.yticks(range(len(corr.columns)), corr.columns)
   # show plot
   plt.show()
```

### ML Pipeline

#### Performing one hot encoding, and normalizing the data if necessary

```python
transactions_master_df.columns
```

Defining features and target using the dataframe we have

```python
target = transactions_master_df['price']
features = transactions_master_df[['document_type', 'transaction_type', 'recording_year',
       'recording_month', 'recording_day', 'contract_day', 'area_sqft',
       'built_year', 'lot_size', 'no_of_baths', 'no_of_bedrooms',
       'property_type', 'tax_amount']]
# ADD CITY AND ZIP_CODE AS WELL, RIGHT NOW THEY ARE REMOVED
features.head()
```

Splitting the data to train and test data

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=21)
```

Here we will create a ColumnTransformer, using advanced features: we use a mask to select the column subsets based on the dtypes, and we use another pipeline to combine imputation and scaling for the numerical features (if necessary). And we also integrate the one hot encoding of categorical features within this pipeline

```python
categorical_features = (features.dtypes == 'object')
numerical_features = ~categorical_features
```

```python
categorical_features
```

```python
numerical_features
```

```python
hot_encoder = OneHotEncoder(handle_unknown='ignore')
```

```python
# One hot encoder has the setting so it ignores any new categories introduced in test data
preprocess = make_column_transformer(
    (hot_encoder, categorical_features))
```

```python
#preprocess.fit_transform(features)
```

```python
linear_model = LinearRegression()
```

```python
pipeline = make_pipeline(
    preprocess,
    linear_model)
```

```python
pipeline.fit(X_train, y_train)
```

### Model Evaluation
